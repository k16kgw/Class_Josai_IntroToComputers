```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

# ====  ====
# OSIPAex
# Windows: "MS Gothic" 
# macOS:   "Hiragino Sans" 
# rcParams['font.family'] = 'IPAexGothic'  # 
rcParams['font.family'] = 'Hiragino Sans'  # 

# ====  ====
t = np.linspace(0, 4*np.pi, 500)      # 
analog = np.sin(t) + 1               # 02 +1
digital = np.where(analog >= 1, 2, 0) # 1210

# ====  ====
plt.figure(figsize=(10, 6))

# 
plt.plot(t, analog, label='', color='blue')

# 
plt.step(t, digital, where='mid', label='', color='red')

# 
plt.xlabel(' [s]')
plt.ylabel('')
plt.title('')

# 
plt.grid(True)
plt.legend()

plt.show()
```
